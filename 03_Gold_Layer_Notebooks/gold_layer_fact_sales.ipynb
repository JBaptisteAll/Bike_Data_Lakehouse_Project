# Init

```python
import pyspark.sql.functions as F
from pyspark.sql.window import Window
```

# Read from Silver Table/s

```python
query = """
SELECT 
  *
FROM data_lakehouse_project.silver.crm_sales
"""
df = spark.sql(query)
```

```python
display(df)
```

# Write into Gold Layer

```python
(
    df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("data_lakehouse_project.gold.fact_sales")
)
```
