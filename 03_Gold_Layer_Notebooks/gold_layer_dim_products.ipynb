# Init

```python
import pyspark.sql.functions as F
from pyspark.sql.window import Window
```

# Read from Silver Table/s

```python
query = """
SELECT 
  prd.product_id,
  prd.category_id,
  prd.product_key,
  px.category,
  px.sub_category,
  prd.product_name,
  prd.product_cost,
  prd.product_line,
  prd.product_start_date,
  prd.product_end_date,
  px.maintenance
FROM data_lakehouse_project.silver.crm_products AS prd
LEFT JOIN data_lakehouse_project.silver.erp_px_cat_g1v2 AS px
  ON px.id = prd.category_id
"""

df = spark.sql(query)
```

```python
display(df)
```

# Write into Gold Layer

```python
(
    df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("data_lakehouse_project.gold.dim_products")
)
```
