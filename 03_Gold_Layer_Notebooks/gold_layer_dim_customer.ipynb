# Init

```python
import pyspark.sql.functions as F
from pyspark.sql.window import Window
```

# Read from Silver Table/s

```python
query = """
SELECT 
  crm_c.customer_id,
  crm_c.customer_key,
  crm_c.firstname,
  crm_c.lastname,
  crm_c.marital_status,
  erp_cust.birthdate,
  erp_loc.country,
  crm_c.gender,
  crm_c.create_date
FROM data_lakehouse_project.silver.crm_customers AS crm_c
LEFT JOIN data_lakehouse_project.silver.erp_loc_a101 AS erp_loc 
  ON crm_c.customer_key = erp_loc.customer_key
LEFT JOIN data_lakehouse_project.silver.erp_cust_az12 AS erp_cust
  ON crm_c.customer_key = erp_cust.customer_key
"""
df = spark.sql(query)
```

```python
display(df)
```

# Write into Gold Layer

```python
(
    df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("data_lakehouse_project.gold.dim_customers")
)
```
