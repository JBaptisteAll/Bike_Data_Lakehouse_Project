# Reading from CSV file

## Source CRM

```python
%python
df_customer = (spark.read.option("header", True)
  .option("inferSchema", True)
  .csv("/Volumes/data_lakehouse_project/bronze/source_system/source_crm/cust_info.csv"))

df_produit = (spark.read.option("header", True)
  .option("inferSchema", True)
  .csv("/Volumes/data_lakehouse_project/bronze/source_system/source_crm/prd_info.csv"))

df_sales = (spark.read.option("header", True)
  .option("inferSchema", True)
  .csv("/Volumes/data_lakehouse_project/bronze/source_system/source_crm/sales_details.csv"))
```

## Source ERP

```python
df_erp_cust = (spark.read.option("header", True)
  .option("inferSchema", True)
  .csv("/Volumes/data_lakehouse_project/bronze/source_system/source_erp/CUST_AZ12.csv"))

df_loc = (spark.read.option("header", True)
  .option("inferSchema", True)
  .csv("/Volumes/data_lakehouse_project/bronze/source_system/source_erp/LOC_A101.csv"))

df_px_cat = (spark.read.option("header", True)
  .option("inferSchema", True)
  .csv("/Volumes/data_lakehouse_project/bronze/source_system/source_erp/PX_CAT_G1V2.csv"))
```

# Writing in to The Bronze layer

## Source CRM

```python
%python
df_customer.write.mode("overwrite").saveAsTable("data_lakehouse_project.bronze.crm_cust_info")
df_produit.write.mode("overwrite").saveAsTable("data_lakehouse_project.bronze.crm_prd_info")
df_sales.write.mode("overwrite").saveAsTable("data_lakehouse_project.bronze.crm_sales_details")
```

## Source ERP

```python
df_erp_cust.write.mode("overwrite").saveAsTable("data_lakehouse_project.bronze.erp_cust_az12")
df_loc.write.mode("overwrite").saveAsTable("data_lakehouse_project.bronze.erp_loc_a101")
df_px_cat.write.mode("overwrite").saveAsTable("data_lakehouse_project.bronze.erp_px_cat_g1v2")
```
